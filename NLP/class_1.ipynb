## Text Cleaning

```python
text = "Text preprocessing is an extremely important step for many NLP tasks (like classifying/categorizing texts). It's ensuring your data is clean and ready, isn't it? Running and jumping are fun activities."
```

```python
# convert to the lower case

text = text.lower()
print(text)
```

```python
# regex

import re
```

```python
# removing special characters

pattern = r'[^A-Za-z0-9 ]'
text = re.sub(pattern, "", text)
print(text)
```

### Stopword

```python
stopwords = set(".?,!")
stopwords

text = "Text preprocessing is an extremely important step for many NLP tasks (like classifying/categorizing texts). It's ensuring your data is clean and ready, isn't it? Running and jumping are fun activities."

for i in stopwords:
    text = text.replace(i, "")
    
print(text)
```

```python
# removing stopwords

text = "Text preprocessing is an extremely important step for many NLP tasks (like classifying/categorizing texts). It's ensuring your data is clean and ready, isn't it? Running and jumping are fun activities."

pattern = r'[!.,?]'
text = re.sub(pattern, "", text)
print(text)
```

### POS Tagging

```python
# POS - Parts of Speech
```

```python
nltk.download("averaged_perceptron_tagger")
```

```python
nltk.download('averaged_perceptron_tagger_eng')
```

```python
tags = nltk.pos_tag(tokens)
print(tags)
```

v -> vocabulary




```python
def clean_tag(tag):
    if (tag.startswith('V')):
        return 'v'
    elif (tag.startswith('J')):
        return 'a'
    elif (tag.startswith('R')):
        return 'r'
    else:
        return 'n'
```

```python
lemmetised_tokens = []

for token,tag in tags:
    new_tag = clean_tag(tag)
    lemmetised_tokens.append(lem.lemmatize(token, new_tag))
    
print(lemmetised_tokens)
```
