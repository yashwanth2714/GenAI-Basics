```python
import os
from huggingface_hub import InferenceClient
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
hf_api_key = os.getenv('HF_API_KEY')
```

```python
client = InferenceClient(api_key=hf_api_key)
```

```python
response = client.chat.completions.create(
    model='meta-llama/Llama-3.1-8B-Instruct',
    messages=[{'role': 'user', 'content': 'What is the most popular programming language for data science?'}]
)
```

```python
print(response.choices[0].message.content)
```

```python
from transformers import pipeline
```

```python
# pipeline is a class. It provides a uniform framework to use the hugging face models for different use cases. It runs locally.

generator = pipeline(
    task='text-generation',
    model='openai-community/gpt2',
    framework='pt',
    device_map='auto'
)
```

```python
output = generator(text_inputs='What is the most popular programming language for data science?', max_new_tokens=50, temperature = 0.9)
```

```python
print(output[0]['generated_text'])
```

```python
smp = pipeline(
    task='sentiment-analysis',
    model='distilbert/distilbert-base-uncased-finetuned-sst-2-english',
    framework='pt',
    device_map='auto'
)
```

```python
res = smp(inputs='The food was really bad')
```
